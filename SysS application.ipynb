```python
import numpy as np
```

```python
vocab = open('/home/a/Downloads/ACL2012_wordVectorsTextFile/vocab.txt', 'rb').read().split()
```

```python
vectors = open('/home/a/Downloads/ACL2012_wordVectorsTextFile/wordVectors.txt', 'rb').read().rsplit('\n')
```

```python
wordlist = open('/home/a/Downloads/500-worst-passwords.txt', 'rb').read().split()
```

```python
indexlist = []
for word in wordlist:
    for i in range(len(vocab)):
        if vocab[i]==word:
            indexlist.append(i)
```

```python
for i in indexlist:
    print(vocab[i])
```

```python
for i in indexlist:
    print(vectors[i])
```

```python
targetvectors = []
for i in indexlist:
    targetvectors.append(np.asarray(vectors[i].split()))
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE()
```

```python
targetvectors[0]
```

```python
wordspace = tsne.fit_transform(targetvectors)
```
